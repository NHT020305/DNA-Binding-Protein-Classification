```python
import numpy as np
import matplotlib.pyplot as plt
```

# Amino Acid Distribution

```python
# Define the classifiers and metrics
classifiers = ["LR", "NB", "KNN", "SVM", "DT", "RF"]
metrics = ["Accuracy", "Sensitivity", "Specificity", "MCC", "AUC"]

# Data for each classifier with Full AAD and Selected AAD
full_aad = [
    [0.64, 0.64, 0.64, 0.27, 0.70],  # LR
    [0.65, 0.74, 0.60, 0.32, 0.74],  # NB
    [0.59, 0.92, 0.42, 0.34, 0.80],  # KNN with k = 55
    [0.72, 0.74, 0.71, 0.43, 0.80],  # SVM
    [0.69, 0.75, 0.66, 0.39, 0.78],  # DT
    [0.72, 0.74, 0.71, 0.43, 0.81]   # RF
]

selected_aad = [
    [0.67, 0.18, 0.92, 0.16, 0.69],  # LR
    [0.67, 0.69, 0.66, 0.33, 0.74],  # NB
    [0.71, 0.58, 0.79, 0.36, 0.75],  # KNN k = 187
    [0.70, 0.72, 0.72, 0.41, 0.78],  # SVM
    [0.68, 0.79, 0.62, 0.39, 0.78],  # DT
    [0.71, 0.79, 0.67, 0.43, 0.81],  # RF
]

tf_idf = [
    [0.73, 0.63, 0.66, 0.34, 0.75],  # LR
    [0.63, 0.62, 0.62, 0.24, 0.62],  # NB
    [0.64, 0.75, 0.71, 0.38, 0.76],  # KNN with k = 37
    [0.74, 0.68, 0.70, 0.40, 0.78],  # SVM
    [0.55, 0.68, 0.64, 0.22, 0.61],  # DT
    [0.54, 0.82, 0.73, 0.37, 0.76]   # RF
]

# Convert data into arrays for easier manipulation
full_aad = np.array(full_aad)
selected_aad = np.array(selected_aad)
tf_idf = np.array(tf_idf)

# Plotting
x = np.arange(len(metrics))  # label locations for each metric
width = 0.25  # width of the bars

fig, axes = plt.subplots(2, 3, figsize=(18, 10), sharey=True)
fig.suptitle('Comparative Bar Graph of Classifier Performance on Full AAD vs Selected AAD', fontsize=18)

for i, ax in enumerate(axes.flat):
    # Extract data for each classifier
    full_data = full_aad[i]
    selected_data = selected_aad[i]
    tf_idf_data = tf_idf[i]

    # Plot Full AAD and Selected AAD data as grouped bars for each classifier
    bars1 = ax.bar(x - width * 1.1, full_data, width, label='Full AAD')
    bars2 = ax.bar(x, selected_data, width, label='Selected AAD')
    bars3 = ax.bar(x + width * 1.1, tf_idf_data, width, label='TD-IDF')

    # Add data labels
    for bar1, bar2, bar3 in zip(bars1, bars2, bars3):
        ax.text(bar1.get_x() + bar1.get_width() / 2, bar1.get_height() + 0.01, f'{bar1.get_height():.2f}', ha='center', va='bottom', fontsize=10)
        ax.text(bar2.get_x() + bar2.get_width() / 2, bar2.get_height() + 0.01, f'{bar2.get_height():.2f}', ha='center', va='bottom', fontsize=10)
        ax.text(bar3.get_x() + bar3.get_width() / 2, bar3.get_height() + 0.01, f'{bar3.get_height():.2f}', ha='center', va='bottom', fontsize=10)

    # Set title, labels, and ticks
    ax.set_title(classifiers[i], fontsize=14)
    ax.set_xticks(x)
    ax.set_xticklabels(metrics, rotation=45)
    ax.set_ylim(0, 1)  # to keep all bars within the same scale
    ax.legend()

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.95])  # leave space for the main title
plt.show()
```

# Deep Learning Models: CNN



```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the table
models = ["Embedding-CNN", "Embedding-ProtCNN", "Integer-CNN", "Integer-ProtCNN"]
sensitivity = [0.28, 0.32, 0.25, 0.31]
specificity = [0.87, 0.89, 0.90, 0.90]
accuracy = [0.68, 0.70, 0.68, 0.70]
mcc = [0.19, 0.26, 0.30, 0.26]
auc_roc = [0.58, 0.61, 0.57, 0.60]

# Plotting
x = np.arange(len(models)) # the label locations
width = 0.15 # width of the bars

fig, ax = plt.subplots(figsize=(12, 7))

# Plot each metric as a set of bars with data labels on top
bars1 = ax.bar(x - 2*width, sensitivity, width, label='Sensitivity')
bars2 = ax.bar(x - width, specificity, width, label='Specificity')
bars3 = ax.bar(x, accuracy, width, label='Accuracy')
bars4 = ax.bar(x + width, mcc, width, label='MCC')
bars5 = ax.bar(x + 2*width, auc_roc, width, label='AUC-ROC')

# Add data labels on top of each bar
def add_labels(bars):
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, yval + 0.01, f'{yval:.2f}', ha='center', va='bottom', fontsize = 12)

add_labels(bars1)
add_labels(bars2)
add_labels(bars3)
add_labels(bars4)
add_labels(bars5)

# Adding labels and title
ax.set_xlabel('Models', fontsize = 15)
ax.set_ylabel('Scores', fontsize = 15)
ax.set_title('Comparison of CNN Models Performance across Different Metrics', fontsize = 15)
ax.set_xticks(x)
ax.set_xticklabels(models, fontsize = 13)
ax.legend(fontsize = 11)

# Display the plot
plt.tight_layout()
plt.show()
```

# Physicochemical Properties


```python
# Names of the metrics and models
metrics = ['Accuracy', 'Sensitivity', 'Specificity', 'MCC', 'AUC']
models = ["Logistic Regression", "Naive Bayes", "KNN", "SVM", "Decision Tree", "Random Forest"]

physicochemical_properties = [
    [0.69, 0.62, 0.72, 0.33, 0.73],  # LR
    [0.70, 0.44, 0.83, 0.29, 0.69],  # NB
    [0.64, 0.79, 0.56, 0.33, 0.75],  # KNN with k = 173
    [0.67, 0.74, 0.63, 0.35, 0.75],  # SVM
    [0.67, 0.69, 0.66, 0.33, 0.74],  # DT
    [0.70, 0.61, 0.74, 0.34, 0.74]   # RF
]

# Group the values by metric instead of by model
metric_values = np.array(physicochemical_properties).T

# Plot settings
x = np.arange(len(models))
width = 0.13  # Width of bars

# Initialize plot
fig, ax = plt.subplots(figsize=(12, 6))

# Plot each metric across all models and add labels on bars
for i, metric in enumerate(metrics):
    bars = ax.bar(x + i * width, metric_values[i], width, label=metric)
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width() / 2, yval, f'{yval:.2f}', ha='center', va='bottom', fontsize = 8)

# Labeling
ax.set_xlabel('Models')
ax.set_ylabel('Scores')
ax.set_title('Comparative Bar Graph of Classifier Performance on Full Physicochemical Properties')
ax.set_xticks(x + 2 * width)
ax.set_xticklabels(models)
ax.legend(title='Metrics', loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=5)

plt.tight_layout()
plt.show()
```
